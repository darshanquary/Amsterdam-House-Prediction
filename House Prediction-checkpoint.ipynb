```python
def convert(value):
    if isinstance(value, (int, float)):
        return float(value)

    temp = str(value).split('-')

    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1])) / 2

    return float(temp[0])
    
```

```python
df['Area'] = df['Area'].apply(convert)
```

```python
df['price_per_area'] = df['Price'] / df['Area']
```

```python
df['Zip'] = df['Zip'].apply(lambda x: x.strip()) 
Address_count = df['Zip'].value_counts()
```

```python
zip_count_less_1 = Address_count[Address_count<=1] 
zip_count_less_1
```

```python
df['Zip'] = df['Zip'].apply(lambda x: 'other' if x in zip_count_less_1 else x)
```

```python
def Room_outlier_remover(df):
    exclude_indices = np.array([])

    for location, location_df in df.groupby('Zip'):
        bhk_stats = {}

        for bhk, bhk_df in location_df.groupby('Room'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_area),
                'std': np.std(bhk_df.price_per_area),
                'count': bhk_df.shape[0]
            }

        for bhk, bhk_df in location_df.groupby('Room'):
            stats = bhk_stats.get(bhk-1)

            if stats and stats['count'] > 5:
                exclude_indices = np.append(
                    exclude_indices,
                    bhk_df[bhk_df.price_per_area < stats['mean']].index.values
                )

    return df.drop(exclude_indices, axis='index')
```

```python
df = Room_outlier_remover(df)
```

```python
df = df.drop(columns=['price_per_area'])
```

```python
df = df.dropna(subset=['Zip','Area','Room','Price'])
```

```python
X = df.drop(columns=['Price']) 
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Linear Regg apply

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown='ignore',sparse_output=False),['Zip']), 
                                       remainder='passthrough')
```

```python
scaler = StandardScaler() 
```

```python
ir = LinearRegression() 
```

```python
pipe = make_pipeline(column_trans,scaler,ir)
```

```python
print(y.isnull().sum())
```

```python
df.head()
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ir = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_ir)
```

# Aplly lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test) 
r2_score(y_test,y_pred_lasso)
```

# apply ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test) 
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test, y_pred_ir))
print("Lasso: ", r2_score(y_test, y_pred_lasso))
print("Ridge: ", r2_score(y_test, y_pred_ridge))
```
